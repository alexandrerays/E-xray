# Dataset

```python
df_af_2017 = pd.read_csv('../../data/bases_ale/ideb_municipios_2017_af.csv')
```

```python
df_af_2017.head()
```

```python
df_af_2017['num_professores_por_num_estudantes'] = np.divide(df_af_2017['num_professores'], df_af_2017['num_estudantes'])
df_af_2017['num_funcionarios_por_num_estudantes'] = np.divide(df_af_2017['num_funcionarios'], df_af_2017['num_estudantes'])
df_af_2017['taxa_escolas_em_atividade'] = np.divide(df_af_2017['num_escolas_em_atividade'], df_af_2017['num_escolas'])
```

```python
df_af_2017['ideb'].value_counts()
```
